```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(f'../results/15wt_results.csv', header=None)
df.columns = ["node", "broadcast_time", "seen_id", "told_by", "hop", "start", "end"]
df["diff_time"] = df["end"]-df["start"]
df = df.groupby(["node", "seen_id", "broadcast_time"], as_index=False).sum().drop(["told_by", "hop", "start", "end"], axis=1)
df["frecuency"] = 1/df["broadcast_time"]
sns.lmplot(x="frecuency", y="diff_time", hue="seen_id", col="node", col_wrap=2, data=df, logx=True, scatter=False, ci=0);
```
